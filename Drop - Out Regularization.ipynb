```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/imamanmehrotra/Datasets/main/sonar_csv.csv', header=None)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

y=df.iloc[:,-1]
X=df.iloc[:,:-1]
```

```python
y=pd.get_dummies(y,drop_first=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7, random_state=42)
X_train.shape, y_train.shape, X_test.shape,y_test.shape
```

```python
model=keras.Sequential([
keras.layers.Dense(60,input_dim=60,activation='relu'),
keras.layers.Dense(30,activation='relu'),
keras.layers.Dense(15,activation='relu'),
keras.layers.Dense(1,activation='sigmoid')
])

model.compile(loss='binary_crossentropy', optimizer='adam',metrics=['accuracy'])

model.fit(X_train, y_train,epochs=100, batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test).reshape(-1)
y_pred=np.round(y_pred)
y_pred=pd.DataFrame(y_pred, columns=['Y_pred']).astype(int)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, classification_report, plot_confusion_matrix,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred), annot=True, cmap='summer');
```
